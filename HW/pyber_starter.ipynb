```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
joined_df = pd.merge(city_df, ride_df, on = 'city', how='inner')

# Display the data table for preview
joined_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
#Create frames for different city types
urban_df = joined_df.loc[(joined_df['type'] == 'Urban'), :]
suburban_df = joined_df.loc[(joined_df['type'] == 'Suburban'), :]
rural_df = joined_df.loc[(joined_df['type'] == 'Rural'), :]
```

```python
# Obtain the x and y coordinates for each of the three city types
urban_x_axis = np.arange(len(urban_df))
urban_y_axis = urban_df['fare']
urban_tick_locations = [value for value in urban_x_axis]

suburban_x_axis = np.arange(len(suburban_df))
suburban_y_axis = suburban_df['fare']
suburban_tick_locations = [value for value in suburban_x_axis]

rural_x_axis = np.arange(len(rural_df))
rural_y_axis = rural_df['fare']
rural_tick_locations = [value for value in rural_x_axis]

# Build the scatter plots for each city types
urbanscatter = plt.scatter(urban_x_axis, urban_y_axis, marker="o", facecolors="red", edgecolors="black",
            s=urban_x_axis, alpha=0.25)

suburbanscatter = plt.scatter(suburban_x_axis, suburban_y_axis, marker="o", facecolors="green", edgecolors="black",
            s=suburban_x_axis, alpha=0.25)

ruralscatter = plt.scatter(rural_x_axis, rural_y_axis, marker="o", facecolors="blue", edgecolors="black",
            s=rural_x_axis, alpha=0.25)

# Incorporate the other graph properties
plt.title("Fare Amount vs City Type")
plt.xlabel("City Type")
plt.ylabel("Fare Amount")

# Create a legend
plt.legend(loc="upper right")

# Incorporate a text label regarding circle size

# Save Figure
plt.tight_layout()
plt.savefig("../Images/pyber.png")
plt.show()
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
grouped_df = joined_df.groupby(['type'])

type_counts = joined_df['type'].value_counts()
fare_sum = grouped_df['fare'].sum()

merged_table = pd.DataFrame({'Number of Fares': type_counts,'Total Fare $': fare_sum})

merged_table.head()

# Build Pie Chart
# fare_sum.plot(kind='pie', subplots=True)
plt.pie(fare_sum, autopct="%1.1f%%", shadow=True, startangle=140)
# Save Figure
plt.savefig("../Images/pyber2.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
# total_rides = grouped_df['rideid'].count()
ride_counts = joined_df['ride_id'].count()
ride_table = pd.DataFrame({'Number of Rides': type_counts})
ride_table.head()

# Build Pie Chart
ride_table.plot(kind='pie', subplots=True)
# Save Figure
plt.savefig("../Images/pyber3.png")
```

```python
# Show Figure
plt.show()
```
