```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "Resources/purchase_data.csv"

# Read purchasing file and store into pandas data frame
df = pd.read_csv(file_to_load)
df.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.
* Create a summary data frame to hold the results
* Optional: give the displayed data cleaner formatting
* Display the summary data frame

```python
#calculate the number of unique items
items = len(df["Item Name"].unique())

#calculate the average price
price = df["Price"].mean()

#number of male players
male_df = df.loc[df['Gender'] == "Male"]
male = len(male_df['SN'].unique())

#number of female players
female_df = df.loc[df['Gender'] == "Female"]
female = len(female_df['SN'].unique())

#number of other/undisclosed players
und_df = df.loc[df['Gender'] == "Other / Non-Disclosed"]
und = len(und_df['SN'].unique())

#calculate the average age
age = df["Age"].mean()

summary_table = pd.DataFrame({"Unique Items": [items],
                              "Average Price": [price],
                              "# of Male Players": male,
                              "# of Female Players": female,
                              "# of Und/Non": und,
                              "Average Age": age})

summary_table
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Total # of players
total = df['SN'].count()

#number/percent of male players
male_df = df.loc[df['Gender'] == "Male"]
male_num = male_df['Gender'].count()
male_pct = (male_num / total) * 100

#number of female players
female_df = df.loc[df['Gender'] == "Female"]
female_num = female_df['Gender'].count()
female_pct = (female_num / total) * 100

#number of other/undisclosed players
und_df = df.loc[df['Gender'] == "Other / Non-Disclosed"]
und_num = und_df['Gender'].count()
und_pct = (und_num / total) * 100

summary_table = pd.DataFrame({"Male": [male_num],
                              "Male %": [male_pct],
                              "Female": [female_num],
                              "Female %": [female_pct],
                              "Undisclosed": [und_num],
                              "Undisclosed %": [und_pct]})

summary_table
```
